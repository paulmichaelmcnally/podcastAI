```python
#INTRO
#THE GOOGLE NEWS

import requests
from bs4 import BeautifulSoup
from datetime import datetime, timedelta

# Set up Google News URL
google_news_url_today = "https://news.google.com/rss/search?q=Johannesburg%20South%20Africa%20when:1d&hl=en-US&gl=US&ceid=US:en"
google_news_url_yesterday = "https://news.google.com/rss/search?q=Johannesburg%20South%20Africa%20when:2d-1d&hl=en-US&gl=US&ceid=US:en"

# Get the current date and time
current_date = datetime.utcnow()

# Subtract one day from the current date to get yesterday's date
yesterday_date = current_date - timedelta(days=1)

# Send the request to Google News for today's news
google_news_response_today = requests.get(google_news_url_today)

# Parse the response from Google News
google_news_data_today = BeautifulSoup(google_news_response_today.text, "xml")

# Extract the stories from Google News that were published today
google_news_stories_today = []
for story in google_news_data_today.find_all("item"):
    if datetime.strptime(story.pubDate.text, "%a, %d %b %Y %H:%M:%S %Z").date() == current_date.date():
        google_news_stories_today.append({
            'title': story.title.text,
            'date': story.pubDate.text,
            'description': story.description.text,
            'link': story.link.text
        })

# Send the request to Google News for yesterday's news
google_news_response_yesterday = requests.get(google_news_url_yesterday)

# Parse the response from Google News
google_news_data_yesterday = BeautifulSoup(google_news_response_yesterday.text, "xml")

# Extract the stories from Google News that were published yesterday
google_news_stories_yesterday = []
for story in google_news_data_yesterday.find_all("item"):
    if datetime.strptime(story.pubDate.text, "%a, %d %b %Y %H:%M:%S %Z").date() == yesterday_date.date():
        google_news_stories_yesterday.append({
            'title': story.title.text,
            'date': story.pubDate.text,
            'description': story.description.text,
            'link': story.link.text
        })

# Extract the top eight stories from today's news
top_stories = google_news_stories_today[:8]

# Loop through the top stories and retrieve their full text content
for i, story in enumerate(top_stories):
    story_title = story['title']
    story_date = story['date']
    story_description = story['description']
    story_link = story['link']

    # Get the content of the story
    try:
        story_response = requests.get(story_link)
        story_data = BeautifulSoup(story_response.text, "html.parser")
        story_text = story_data.get_text()
    except:
        # Skip stories with errors or empty text
        continue

    # Skip stories with empty or None text
    if not story_text:
        continue

    # Store the story details and text in a dictionary
    story_data = {
        'title': story_title,
        'date': story_date,
        'description': story_description,
        'link': story_link,
        'text': story_text
    }

    # Print the story details
    print(f"{i+1}. Title: {story_title}")
    print(f"   Date: {story_date}")
    print(f"   Description: {story_description}")
    print(f"   Link: {story_link}\n")

# Ask the user to select a story
while True:
    try:
        selection = int(input("Please enter the number of the story you want to use: "))
        if selection < 1 or selection > len(top_stories):
            print("Invalid selection. Please enter a number between 1 and", len(top_stories))
        else:
            break
    except ValueError:
        print("Invalid input. Please enter a number.")

# Get the selected story details
selected_story = top_stories[selection-1]
selected_story_link = selected_story['link']
selected_story_title = selected_story['title']
selected_story_date = selected_story['date']
selected_story_description = selected_story['description']

# Get the content of the selected story
try:
    selected_story_response = requests.get(selected_story_link)
    selected_story_data = BeautifulSoup(selected_story_response.text, "html.parser")
    selected_story_text = selected_story_data.get_text()
except:
    # Set the selected story text to None if there's an error
    selected_story_text = None

# Store the selected story details and text in a dictionary
top_story_data1 = {
    'title': selected_story_title,
    'date': selected_story_date,
    'description': selected_story_description,
    'link': selected_story_link,
    'text': selected_story_text
}

# Print the selected story details and text
print(f"\nSelected Story: {selected_story_title}")
print(f"Date: {selected_story_date}")
print(f"Description: {selected_story_description}")
print(f"Link: {selected_story_link}\n")

if selected_story_text:
    print("Selected Story Text:")
    print("---------------------------------------------------------------------")
    print(selected_story_text)
    print("\n\n")
else:
    print("No text available for this story.\n")
```

```python
#INTRO
#NORMALISE THE VOLUME

import os
from pydub import AudioSegment

# Define a target loudness level in dBFS (decibels relative to full scale)
target_loudness = -20

# Define the directory where the MP3 files are located
directory = "/content/"

# Loop through all the files in the directory
for filename in os.listdir(directory):
    # Check if the file starts with "introline"
    if filename.startswith("introline"):
        # Extract the number from the original file name
        file_number = filename.split("introline")[1].split(".")[0]

        # Process the file
        print(f"Processing file: {filename}")
        # Load the audio file
        audio = AudioSegment.from_file(os.path.join(directory, filename), format="mp3")

        # Calculate the current loudness level
        current_loudness = audio.dBFS

        # Calculate the difference in loudness between the current level and the target level
        loudness_difference = target_loudness - current_loudness

        # Apply the normalization gain to the audio
        normalized_audio = audio.apply_gain(loudness_difference)

        # Export the normalized audio as an MP3 file
        normalized_file_name = f"introlineloud{file_number}.mp3"
        normalized_audio.export(os.path.join(directory, normalized_file_name), format="mp3")

        print(f"Exported file: {normalized_file_name}")
```

```python
#INTRO
#COMBINE THE INTRO MP3S

import os
from pydub import AudioSegment
from google.colab import files

# Set the path to the directory where the MP3 files are stored
dir_path = '/content/'

# Set the output file name
output_file_name = 'intro_combined_loud.mp3'

# Get a list of all the MP3 files in the directory
mp3_files = [f for f in os.listdir(dir_path) if f.startswith('introlineloud') and f.endswith('.mp3')]

# Sort the files numerically based on the number in the file name
mp3_files.sort(key=lambda f: float('inf') if not f.split('.')[0].split('introlineloud')[1].isdigit() else int(f.split('.')[0].split('introlineloud')[1]))

# Concatenate the MP3 files
combined_audio = AudioSegment.empty()
for file in mp3_files:
    file_path = os.path.join(dir_path, file)
    audio_segment = AudioSegment.from_file(file_path, format='mp3')
    combined_audio += audio_segment
    print(f"Concatenating file: {file}")

# Export the final output as an MP3 file
combined_audio.export(output_file_name, format='mp3')

# Download the output file
files.download(output_file_name)
```
